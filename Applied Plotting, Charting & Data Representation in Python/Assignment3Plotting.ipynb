In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
df1 = df.T
df1 = df1.describe()
df1 = df1[0:-5] 
df1
```

```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors

Mean =  df.mean(axis=1) 
Std_Err = df.std(axis=1)/np.sqrt(df.shape[1]) # Error estandar de la media.
Niv_Conf = Std_Err*1.96 # Usamos el 1.96 como coeficiente alpha para obtener un nivel de confianza del 95% siando un valor usual en estadística

Range_Lines = []


fig, ax = plt.subplots()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
Bar = plt.bar(range(len(Mean)), Mean, yerr= Niv_Conf, color = '#4d94ff') 
plt.xticks(range(len(Mean)), df.index)
plt.title('Task-Driven Visualization', fontsize=17)
Colors = plt.get_cmap('viridis') 
Normalize = mpl.colors.Normalize(vmin=0, vmax=2)

def onclick(event):
    global Range_Lines
    global Bar
    
    #Addign a horizontal line with the first click where clicked
    if len(Range_Lines) == 0:
        Range_Lines.append(event.ydata)
        plt.axhline(y=event.ydata, color = 'black', linestyle = '-')
       # y_text = plt.text(1995.9, event.ydata, 'y = %d' %Y, bbox=dict(fc='white', ec='k'));
    # With the second click, add the second horizontal line and giving color the area between both lines
    elif len(Range_Lines) == 1:
        Range_Lines.append(event.ydata)
        Range_Lines.sort()
        URange = Range_Lines[1]
        DRange = Range_Lines[0]
        plt.axhline(y=event.ydata, color = 'black', linestyle = '-')
        plt.axhspan(DRange, URange, facecolor='aquamarine', alpha = 0.5)
        
        #Programing the change of color in the Bars
        for i, bar in enumerate(Bar):
            ULim = bar.get_height() + Niv_Conf.iloc[i] #Higher lim in the bar
            DLim = bar.get_height() - Niv_Conf.iloc[i] #Lower lim in the bar
            
            #selecting the color with high ranges
            if DRange > ULim:
                bar.set_color(Colors(Normalize(0)))
                
            #selecting the color with low ranges
            elif URange < DLim:
                bar.set_color(Colors(Normalize(2)))
                
            # When range is included in the distribution, setting in  white
            elif URange > ULim and DRange < DLim:
                bar.set_color(Colors(Normalize(1)))
                
            # When range is included in the bar range, calculating the overlap area and color for the bar
            elif URange < ULim and DRange > DLim:
                Overlap = (URange - DRange) / (Ulim - Dlim)
                bar.set_color(Colors(Normalize(Overlap)))
                
            # When range is partially included in the bar range slightly high, calculating the overlap area and color for the bar
            elif DRange < ULim < URange:
                Overlap = (ULim - DRange) / (Ulim - Dlim)
                bar.set_color(Colors(Normalize(Overlap)))
                
            # When range is partially included in the bar range slightly low, calculating the overlap area and color for the bar
            elif URange > DLim > DRange:
                Overlap = (URange - DLim) / (Ulim - Dlim)
                bar.set_color(Colors(Normalize(Overlap)))
                    
    else:
        Range_Lines = []
        plt.cla()
        Bar = plt.bar(range(len(Mean)), Mean, yerr = Niv_Conf, color = '#4d94ff')
        plt.xticks(range(len(Mean)), df.index)
        plt.title('Task-Driven Visualization', fontsize = 17)
    
scmppl = cm.ScalarMappable(norm=Normalize, cmap=Colors)
scmppl.set_array(np.arange(-1,1,0.1))
Col_Bar = plt.colorbar(scmppl, ticks=[0, 1, 2])
Col_Bar.ax.set_yticklabels(['0', '1','0'])
plt.show()

_ = plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```
