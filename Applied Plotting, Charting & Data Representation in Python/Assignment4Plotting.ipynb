```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
from mpl_toolkits.mplot3d import axes3d

%matplotlib notebook
```

```python
sni = pd.read_csv('Investigadores-SNI-Vigentes-2018.csv')
sni = pd.DataFrame(sni)
sni
```

```python
snin = sni.sort_values(by = 'Nivel')
sniC = snin[snin['Nivel']=='C'].shape[0]
sni1 = snin[snin['Nivel']=='1'].shape[0] 
sni2 = snin[snin['Nivel']=='2'].shape[0] 
sni3 = snin[snin['Nivel']=='3'].shape[0] 
Nivel = ('1', '2', '3','C')
NI = (sni1,sni2, sni3, sniC)

fig = plt.figure().add_axes([0,0,1,1])
fig.bar(Nivel,NI, color= ['darkred','indianred','lightcoral','brown'])
plt.title('Investigadores por Nivel en SNI-2018', fontsize = 17)
plt.xlabel('Nivel SNI')
plt.ylabel('No. de Investigadores')
fig.spines['top'].set_visible(False)
fig.spines['right'].set_visible(False)
plt.savefig('Investigadores_por_nivel_SNI.png', bbox_inches='tight')
plt.show()
```

```python
sni = sni.sort_values(by = 'Institución de Adscripción')
Inst = sni['Institución de Adscripción'].values.tolist()
Inst = np.array(Inst)
G = {elem : pd.DataFrame for elem in Inst}
val = []
Insti=[]
for key in G.keys():
    G[key] = sni[:][sni['Institución de Adscripción']==key]
    L = G[key].shape[0]
    if L >= 500:
        val.append(L)
    if L >= 500:
        Insti.append(key)
for item in Insti:
    print(item)

Instituciones=np.array(['BUAP', 'CINVESTAV', 'IPN', 'ITESM', 'S.I.', 'UANL', 'UAEM', 'UAM', 'UdG', 'UG', 'UNAM' ])    
NI = val


fig = plt.figure().add_axes([0,0,1,1])
fig.bar(Instituciones,NI, color=['gainsboro', 'lightgray', 'silver', 'silver', 'darkgrey', 'darkgray', 'grey', 'gray', 'dimgrey', 'dimgray', 'black'])
plt.title('Investigadores por Institución-SNI-2018', fontsize = 17)
plt.xlabel('Institucion')
plt.ylabel('No. de Investigadores')
fig.spines['top'].set_visible(False)
fig.spines['right'].set_visible(False)
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.savefig('Investigadores_por_Institución_SNI.png', bbox_inches='tight')
plt.show()
```

```python
sni = sni.sort_values(by = 'Área del Conocimiento')
Areas = sni['Área del Conocimiento'].values.tolist()
Areas = np.array(Areas)
R = {elem : pd.DataFrame for elem in Areas}
val = []
Area=[]
for key in R.keys():
    R[key] = sni[:][sni['Área del Conocimiento']==key]
    L = R[key].shape[0]
    val.append(L)
    Area.append(key)
for item in Area:
    print(item)

AreaC=np.array(['Área I', 'Área II', 'Área III', 'Área IV', 'Área V.', 'Área VI', 'Área VII'])    
NI = val


fig = plt.figure().add_axes([0,0,1,1])
fig.bar(AreaC,NI, color=['lightblue', 'paleturquoise', 'darkturquoise', 'mediumturquoise', 'lightseagreen', 'cadetblue', 'darkcyan'])
plt.title('Investigadores por Área-SNI-2018', fontsize = 17)
plt.xlabel('Área del Conocimiento')
plt.ylabel('No. de Investigadores')
fig.spines['top'].set_visible(False)
fig.spines['right'].set_visible(False)
x = plt.gca().xaxis
for item in x.get_ticklabels():
    item.set_rotation(45)
plt.savefig('Investigadores_por_Área_SNI.png', bbox_inches='tight')
plt.show()
```
